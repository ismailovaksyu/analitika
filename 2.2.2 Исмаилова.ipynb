```python
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
url = 'C:\\Users\\пользователь\\anaconda\\iris.data'
iris = np.genfromtxt(url, delimiter = ',', dtype = 'str')
```

```python
iris
```

```python
iris_frame = pd.DataFrame(iris, columns=['длина_ч', 'ширина_ч', 'длина_л', 'ширина_л', 'вид'])
```

```python
iris_frame[['длина_ч', 'ширина_ч', 'длина_л', 'ширина_л']] = iris_frame[['длина_ч', 'ширина_ч', 'длина_л', 'ширина_л']].astype(float)
```

```python
iris_frame[['длина_ч',  'ширина_ч', 'длина_л', 'ширина_л']].mean()
```
